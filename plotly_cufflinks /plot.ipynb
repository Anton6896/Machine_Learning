```python
import pandas as pd
import numpy as np
from plotly.offline import (download_plotlyjs, init_notebook_mode, plot, iplot)
import cufflinks as cf
# for plotting the plots will use the jupiter style scientists desk
```

```python
init_notebook_mode(connected=True)
%matplotlib inline
cf.go_offline()
```

```python
df = pd.DataFrame(
    np.random.randn(100,4),
    columns='A B C D'.split()
)
df.head()
```

```python
df.iplot()
```

```python
df2 = pd.DataFrame(
    {
        'Category' : ' A B C'.split(),
        'values' : [33,43,50]
    }
)
df2.head()
```
